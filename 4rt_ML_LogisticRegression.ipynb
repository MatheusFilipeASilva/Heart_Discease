```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, accuracy_score, ConfusionMatrixDisplay
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.linear_model import LogisticRegression
```

```python
# Tomamos nota que, para modelos com classificação, usar um stratified KFold pode ser uma boa ideia. Assim sendo, usaremos essa abordagem.

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

df = pd.read_csv('df_att.csv')
cat_features = list(df.select_dtypes(exclude='number').drop('Heart Disease', axis=1).columns)
df=pd.get_dummies(df, columns = cat_features, dtype=float)
df['Heart Disease'] = (df['Heart Disease'] == 'Presence').astype(int)
X = df.drop(['Heart Disease', 'Unnamed: 0'], axis=1)
y = df['Heart Disease']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=42, shuffle=True)


scaler = StandardScaler()
oof_preds = np.zeros(len(X_train))
```

```python
for train_idx, val_idx in kf.split(X_train, y_train):
    scaler = StandardScaler()
    X_tr = scaler.fit_transform(X_train.iloc[train_idx])
    X_val = scaler.transform(X_train.iloc[val_idx])
    
    y_tr = y_train.iloc[train_idx]
    y_val = y_train.iloc[val_idx]
    print('ok')

    model = LogisticRegression(
        penalty='elasticnet',
        solver='saga',
        l1_ratio=0.3,
        C=0.1,
        max_iter=5000,
        random_state=42
    )
    model.fit(X_tr, y_tr)
    
    probs = model.predict_proba(X_val)[:, 1]
    oof_preds = probs
```

```python
df_preds = pd.DataFrame({'preds': oof_preds})
df_preds
```
